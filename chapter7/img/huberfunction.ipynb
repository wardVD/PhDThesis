```python
%matplotlib inline

import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [10, 8]


import numpy as np
def getDist(x1,x2,y1,y2):
    return np.sqrt( (x1-x2)**2 + (y2-y1)**2)

def getAngle(x1,x2,y1,y2):
    return np.arctan((y2-y1)/(x2-x1))

def newDist(dist):
    mu = 156
    return np.sqrt(mu*(2*dist - mu))

colors = ['b','c','g','r']

# Plot a line based on the x and y axis value list.
def draw_line():
    
    # List to hold x values.
    x_number_values = [0, 20, 35,170,400]

    # List to hold y values.
    y_number_values = [0, 80, -90,280,-200]

    plt.scatter(x_number_values,y_number_values, )
    # Plot the number in the list and set the line thickness.
    for iother,other in enumerate(x_number_values[1:]):
        plt.plot([x_number_values[0], x_number_values[iother+1]], 
                 [y_number_values[0], y_number_values[iother+1]], 
                 linewidth=3,
                 linestyle='--',
                 color=colors[iother])
        
        dist = getDist(x_number_values[0], x_number_values[iother+1],
                       y_number_values[0], y_number_values[iother+1])
        angle = getAngle(x_number_values[0], x_number_values[iother+1],
                             y_number_values[0], y_number_values[iother+1])
        if dist >= 156:
            dist = newDist(dist)
                
        newX,newY = x_number_values[0]+dist*np.cos(angle), y_number_values[0]+dist*np.sin(angle)
            
        plt.plot([x_number_values[0], newX], 
                     [y_number_values[0], newY], 
                     linewidth=3,
                     linestyle='-',
                     color = colors[iother])
        

    # Set the line chart title and the text font size.
    #plt.title("Square Numbers", fontsize=19)

    # Set x axes label.
    plt.xlabel("Horizontal distance from DOM", fontsize=15)

    # Set y axes label.
    plt.ylabel("Vertical distance from DOM", fontsize=15)

    # Set the x, y axis tick marks text size.
    plt.tick_params(axis='both', labelsize=15)

    # Display the plot in the matplotlib's viewer.
    
    plt.show()
    
    

if __name__ == '__main__':
    draw_line()
```
